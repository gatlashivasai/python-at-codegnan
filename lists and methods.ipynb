#lists
-lists are ordered collection
-lists are mutable
-list contains duplicate elements
-list is heterogenous datatype


```python
#define list
#list is ordered collection
lst=[1,2,4,5,"Hi","Hello",1,3,4]
print(lst)
print(lst[0])
print(lst[1])
print(lst[4])
```

```python
#list allows indexing and slicing
#accesing list items
lst=[1,2,4,5,"Hi","Hello",1,3,4]
print(lst[-4])
lst=[10,20,30,40,50,60,70]
print(lst[:5:-1])
print(lst[:500])
```

```python
#empty list creation
l1=[]
l2=[]
print(type(l1),type(l2))
```

```python
#list operations
#1. list concatination (using + operator)
#2. list repetation (using * operator)
l1=[1,2,3]
l2=[3,4,5,6]
print(l1+l2)
print([0]*10)
```

#llist methods
-item adding methods:
    - append(item): It adds item at end of list
    - extend(iterator):It  adds all the iterator element at end of the list
    - insert(pos,item): It adds item at specific position

```python
lst=[10,20,30,40]
lst.append(100)
print(lst)
lst.extend([1,2,3])
print(lst)
lst.insert(0,1000)=
print(lst)
```

## item removing methods
- remove(item):It removes the item from the list if it exits in the list otherwise value error
- pop(pos): It remove and return the specific position item if position is passed otherwise it remove and returns the last item from list
- clear(): it removes all the elements from the list

```python
lst=[10,20,30,20,40,50,20]
val=lst.pop()
print(val)
print(lst)
lst.clear()
print(lst)
print(lst.remove(30))
print(lst)
```

# index and count()
# index(item) :item is present in list it returns first occurance the index 
# count(item):it returns the total occurance of the items

```python
lst= [10,20,30,20,40,50]
ind=lst.index(20)
print(ind)
print(lst.count(20))
print(lst)
```

```python
# sorting items
lst= [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse=True)
print(lst)
```

```python
# built in functions : len(),min(),max(),sum()
lst=[1,2,3,4,5]
print(len(lst),min(lst),max(lst),sum(lst))
print("average:",sum(lst)/len(lst))
```

```python
#built in functions : len(),min(),max(),sum()
lst=["a","A","Mahi","sanju","Rohit"]
print(len(lst),min(lst),max(lst))
```
