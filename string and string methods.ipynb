#string methods
- Case Conversion Methods
  - upper() : It converts entire string into Upper case
  - lower() : It converts entire string into Lower case
  - title() : It converts the First letter of every word into upper and remaining lower
  -swapcase() :It converts Upper case letter into lower case and lower to upper
  -capitalize() : It converts First character into upper case and remaining into lower case


```python
s="The sky is blue and Ocean is BLUE@123"
print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
```

```python
#split():It splits the entire string into list of substring based on specified 

String="The Sky is blue and Ocean is BLUE@123"
print(String.split("is"))
```

```python
# 'sep'.join(iterable):
words=['The','Sky', 'is','blue','and','ocean','is','blue']
print("".join(words))
print(" ".join(words))
```

```python
#substring index finding
# index(substring): if substring is exists in string it returns the first occurance of substring positon otherwise 'valueError'\
# find(substring):if substing is exists in string it returns the first occurance of substring otherwise,-1
String = "The Sky is blue and Ocean is Blue"
print(String.index('is'))
print(String.index('Z'))
```

```python
String = "The Sky is blue and Ocean is Blue"
print(String.find('is'))
print(String.find('Z'))
print(String.find('e',10))
```

```python
#count(substring): It returns the total non overlapping occurences of substring
s="ABCBCBbbaabb"
print(s.count("B"))
print(s.count("BCB"))
lower_string=s.lower()
print(lower_string)
print(s.count('z'))
```
