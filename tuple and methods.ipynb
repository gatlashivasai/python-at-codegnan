 # Dictionary:
- Dictionary is ordered collection
- key value (i.e items) pairs
- It is mutable and defines with '{}'
# Dictionary format:
{Key:value}
# Characteristics:
- Dictionary is ordered collection
- Dictionary is mutable
- Dictionary does not supporting indexing
- Dictionary keys are unique
- Dictionary values allows duplicates
- List cant be used as dictionary key,because unhashable
- Dictionary values can be accessed using keys

```python
#Dictionary ordered collection
details={"Name":"venky","Course":"python","Batch":57,"age":20}
print(details)
```

```python
#Dictionary values access by using Keys
#updating dictionary values
#adding new elements into dictionary
details={"Name":"venky","Course":"python","Batch":57,"age":20}
print(details['Course'])
#updating
details['age']=25
print(details['age'])
# adding new items
details['location']="hyderabad"
print(details)
details={"age":25,"Name":"venky","Course":"python","Batch":57,"age":20}
print(details)
```

```python
#list cant be used as a dictionary key
details={"Name":"venky","Course":"python",[1,2,3]:57,"age":20}
```

# Dictionary methods
#get(key,default_value:Optional):It returns the value of the key if key is exists in dictionary,otherwise it returns default value

```python
details={"Name":"venky","Course":"python","Batch":57,"age":20}
print(details['Name'])
print(details.get('Name'))
print(details.get('marks'))
print(details.get('marks',"not available"))
```

#update(new_dictionary): This method updates the values if new_dictionary keys alredy exists in dictionary,otherwise it adds new items to the dictionary

```python
details = {'name': 'dsp', 'course': 'python', 'batch': 57, 'age': 22}
details.update({'batch' : 58 , 'course' : 'java' , 'location' : 'hyd'})
print(details)
```

# Keys() : The method returns the list dictionary keys
# Values() : This method returns the list dictionary values
# items(): This method returns the List of tuple of Key-value pairs

```python
print(details.keys())
print(details.values())
print(details.items())
```

# Dictionary remove methods
- popitem() : It return and removes the last item as tuple
- pop(key) : This method removes the specified key item and return the specified keys value,otherwise 'keyerror'

```python
details = {'name': 'dsp', 'course': 'java', 'batch': 58, 'age': 22, 'location': 'hyd'}
item = details.popitem()
print(item)
print(details)
val = details.pop('name')
print(val)
print(details)
```
