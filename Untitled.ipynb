```python
# indexing
s="codegnan"
print(s[2])
print(s[-1])
```

```python
#slicing
s="codegnan"
print(s[2:5])
print(s[2:5:2])
print(s[7:2])
print(s[5:100])
print(s[10:100])
print(s[-4:])
print(s[-2:])
print(s[::-1])
print(s[-3::-1])
print(s[-3::1])
print(s[::-1])
```

```python
#empty string representation
#type() : this function returns the objects datatype
s1=""
s2=str()
print(type(s1))
print(type(s2))
```
